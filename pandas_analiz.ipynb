```python
import pandas as pd
```

```python
dataframe=pd.read_csv('loan_data_set.csv')
```

```python
dataframe.head()
```

```python
dataframe['Gender'].unique()
```

```python
dataframe['Gender'].fillna(dataframe['Gender'].mode()[0],inplace=True)
dataframe['Gender']
```

```python
dataframe['Gender'].unique()
dataframe.dropna(subset='Married',inplace=True)
dataframe.info()
```

```python
dataframe['Dependents'].unique()
dataframe['Dependents'].fillna(dataframe['Dependents'].mode()[0],inplace=True)
dataframe['Self_Employed'].fillna(dataframe['Self_Employed'].mode()[0],inplace=True)
dataframe.info()
```

```python
dataframe['LoanAmount'].fillna(round(dataframe['LoanAmount'].mean(),0),inplace=True)
dataframe['Credit_History'].fillna(dataframe['Credit_History'].mode()[0],inplace=True)
dataframe['Loan_Amount_Term'].fillna(round(dataframe['Loan_Amount_Term'].mean(),0),inplace=True)
dataframe.info()
```

```python
dataframe.drop_duplicates()
dataframe.groupby(['Gender','Married']).count()
```
